```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load company sales data and store to `sales`

```python
company_sales=pd.read_csv('company_sales_data.csv')
company_sales.head()
```

```python
company_sales.info()
```

## Use pandas to impute missing valies if there is any

```python
shirt=company_sales['Shirt'].median()
shirt
```

```python
company_sales['Shirt'].fillna(shirt, inplace=True)
company_sales.info()
```

## Exercise 1.Read Bags, Top, Shirt and Pant sales data and show it  using a multiline plot


```python
fig=plt.figure(figsize=(12,5))
plt.plot(company_sales['Bags'],
        label='Bag Sales Data',
        marker='o')

plt.plot(company_sales['Top'],
        label='Top Sales Data',
        marker='o')

plt.plot(company_sales['Shirt'],
        label='Shirt Sales Data',
        marker='o')

plt.plot(company_sales['Pant'],
        label='Pant Sales Data',
        marker='o')

plt.legend()

plt.suptitle('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

plt.show()
```

## Exercise 2: Read Pant and Shirt sales data of each month and show it using a scatter plot


```python
plt.scatter(company_sales['month_number'], company_sales['Pant'],
            color='red',
           label="Pant Sales Data")

plt.scatter(company_sales['month_number'], company_sales['Shirt'],
            color='blue',
           label='Shirt Sales Data')


plt.legend()

plt.xlabel('Month Number')
plt.ylabel('Number of units sold')

plt.show()
```

## Exercise 3: Read Jeans and Top product sales data and show it using the bar chart

Note:
-  move x axis of jeans .25 left and Top .25 right and add width
-  plt.bar(sales['month_number']-.25, sales['Jeans'],color='red',width=.30, label="Jeans sales data")

```python
plt.bar(company_sales['month_number']-.25, company_sales['Jeans'],
       color='red',
        width=.30,
       label='Jeans sales data')

plt.bar(company_sales['month_number'] +.25, company_sales['Top'],
       color='blue',
        width=.30,
       label='Top sales data')

plt.title('Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

plt.show()
```

## Exercise 4: Read the total profit of each month and show it using the histogram to see the most common profit ranges 
The histogram should look like this:


```python
plt.hist(company_sales['total_profit'],
        label='Profit data')
plt.title('Profit data')
plt.xlabel('Profit in dollar')
plt.legend()

plt.show()
```

## Exercise 5: Calculate `total sale` data for last year for each product and show it using a Pie chart

- salesData=[sales['Bags'].sum(),.......]

```python
salesData=[company_sales['Bags'].sum(),
           company_sales['Pant'].sum(),
           company_sales['Top'].sum(),
           company_sales['Jeans'].sum(),
           company_sales['Shirt'].sum(),
           company_sales['Tie'].sum()]



plt.pie(salesData, labels=['Bags', 'Pant', 'Top', 'Jeans', 'Shirt', 'Tie'], autopct='%.1f')
plt.title('Sales data')

plt.show()
```

### Exercise 6: Read all product sales data and show it using the stack plot. Save the result as `stack_sales.png`


```python
x=company_sales['month_number']
y=np.vstack([company_sales['Bags'],
            company_sales['Tie'],
            company_sales['Shirt'],
            company_sales['Jeans'],
            company_sales['Top'],
            company_sales['Pant']])
labels=['Bags', 'Tie', 'Shirt', 'Jeans', 'Top', 'Pant']
```

```python
plt.stackplot(x, y, labels=labels)
plt.legend(loc='upper left')
plt.show()
```
